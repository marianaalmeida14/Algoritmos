### APP - Registos Médicos 

```python
import matplotlib.pyplot as plt



# Descreve em texto o teu 
# 
# BdEMD = [EMD]
# EMD = [...]
#
# Leitura/carregamento da informaÃ§Ã£o do ficheiro
def getReg(linha):
    linha1=linha.replace("\n", "")
    reg=linha1.split(",")
    return reg


def lerDataset(fnome):
    f = open(fnome, encoding="utf-8")
    bd = []
    f.readline()
    i=1
    for linha in f:
        linhanova=getReg(linha)
        linhanova=linhanova[1:]
        id="emd"+str(i)
        linhanova[0]=id
        bd.append(linhanova)
        i=i+1
    return bd

BD = []
BD = lerDataset("emd.csv")
print(BD)


def chaveOrd(reg):
    return (reg[1])
    
    
def listarDataset(bd):
    print("id   | DataEMD |          Nome         | Resultado ")
    print("__________________________________________________")
    bd.sort(key=chaveOrd, reverse=True)
    for reg in bd:
        string=str(reg[0])+ "|" + str(reg[1])+ "|" + str(reg[2]) + " " + str(reg[3]) + "|" + str(reg[11])            
        print(string)             
BD = []
BD = lerDataset("emd.csv")

listarDataset(BD)


#index 7
def modalidades(bd):
    lista=[]
    for elem in bd:
        if elem[7] not in lista:
            lista.append(str(elem[7])) 
    lista.sort()
    return lista

BD = []
BD = lerDataset("emd.csv")
lista=modalidades(BD)
print(lista)


def distribPorModalidade(bd):
    distribuicaomod={}
    for elem in bd:
        if elem[7] in distribuicaomod.keys():
            distribuicaomod[elem[7]]=distribuicaomod[elem[7]]+1
        else:
            distribuicaomod[elem[7]]=1
    
    return distribuicaomod


BD = []
BD = lerDataset("emd.csv")
lista=distribPorModalidade(BD)
print(lista)

def distribPorClube(bd):
    distribuicaoclub={}
    for elem in bd:
        if elem[8] in distribuicaoclub.keys():
            distribuicaoclub[elem[8]]=distribuicaoclub[elem[8]]+1
        else:
            distribuicaoclub[elem[8]]=1
    
    return distribuicaoclub

BD = []
BD = lerDataset("emd.csv")
lista=distribPorClube(BD)
print(lista)

def distribPorAno(bd):
    distribuicaoano={}
    for elem in bd:
        data=elem[1]
        ano=data[:4]
        if ano in distribuicaoano.keys():
            distribuicaoano[ano]=distribuicaoano[ano]+1
        else:
            distribuicaoano[ano]=1
    
    return distribuicaoano

BD = []
BD = lerDataset("emd.csv")
lista=distribPorAno(BD)
print(lista)

def distrib(bd, i):
    distrib={}
    for elem in bd:
        if elem[i] in distrib.keys():
            distrib[elem[i]]=distrib[elem[i]]+1
        else:
            distrib[elem[i]]=1
    return distrib

BD=[]
BD= lerDataset("emd.csv")
lista=distrib(BD,5)
print(lista)

def plotDistribuicaoPorModalidade(bd):

    c=distribPorModalidade(bd)
    x=c.keys()
    y=[]
    for i in x:
        y.append(c[i])
    plt.bar(x, y)
    plt.show()
    
plotDistribuicaoPorModalidade(BD)

def plotDistrib(d):
    x_axis=[]

    y_axis=[]

    label=[]

    for mod in d.keys():
        y_axis.append(mod)
        x_axis.append(d[mod])
    
    plt.barh(y_axis,x_axis)
    plt.title('Distribuição')
    
    plt.xlabel('Nº de Registos')
    plt.show()

import PySimpleGUI as sg

menu_list_column =[
            [sg.Button("Carregar base de dados")], 
            [sg.Button("Modalidades")],
            [sg.Button("Distribuição por Ano")], 
            [sg.Button("Distribuição por clube")],
            [sg.Button("Listar")]
]
    

data_viewer_column = [
    [sg.Text("Painel de Dados")],
    [sg.Text(size=(40, 1), key="_Dados_")],
]

# ----- Full layout -----
layout = [
    [
        sg.Column(menu_list_column),
        sg.VSeperator(),
        sg.Column(data_viewer_column),
    ]
]

window = sg.Window("App Registos Médicos", layout)   
stop=False

while not stop:
    event, values= window.read()
    
    print(event + ":\n")
    print(values) 
    
    if  event== sg.WIN_CLOSED:
        stop=True
    elif event=="Carregar base de dados":
        cont=0
        for elem in BD:
            cont=cont+1
        window['_Dados_'].Update("Foram carregados " + str(cont)+ " registos.")
        
    elif event=="Modalidades":
        lista=modalidades(BD)
        modalidades=""
        for i in range (len(lista)):
            modalidades=modalidades+ lista[i]    
        window['_Dados_'].Update("Modalidades", modalidades)
                   
            
    elif event == "Distribuicao por ano":
            window['_Dados_'].Update(plotDistrib(distribPorAno(BD)))
    
    elif event == "Distribuicao por clube":
        window['_Dados_'].Update(plotDistrib(distribPorClube(BD)))
    
    elif event == "Listar":
        listar=listarDataset(BD)

        
window.close()
```
