## Resumo:
Este trabalho de casa consiste na implementação do algoritmo BubbleSort integrando-o nas funções previamente elaboradas na aula prática de Algoritmos e Técnicas da Programação.

O algoritmo consiste numa função de ordenação através de trocas diretas.

A ideia é percorrer a lista múltiplas vezes, comparando dois itens adjacente e trocando-os caso estejam num ordem errada.

Deste modo, procedi à criação do ciclo while, que irá parar quando não houver mais trocas.

Com o auxílio do ciclo for e de uma variável temporal, comparei dois elementos consecutivos da lista, trocando-os de ordem caso o segundo valor fosse mais pequeno que o primeiro, e assim sucessivamente.

```python
from random import randint


def menu():
    print(""" Gestão listas de inteiros
          (1)-Introduza a lista "à mão"
          (2)- Gerar lista com valores aleatórios
          (3)- Maior elemento da lista
          (4)- bubble sort
          (0)-Sair
          """)


def criarListaM():
    lista = []
    n = int(input("Introduza o número de elementos da lista:"))
    for i in range(n):
        a = int(input("Introduza o número" + str(i+1) + ":"))
        lista.append(a)
    return lista


def criarListaA():
    n1 = int(input("Introduza o número de elementos da lista:"))
    lista = []
    for i in range(n1):
        element = randint(0, 100)
        lista.append(element)
    return lista


def maiorElemento(lista):
    max = 0
    i = 0
    while i < len(lista):
        if lista[i] > max:
            max = lista[i]
        i = i+1
    return max



def bubbleSort(lista):         
    trocas =1
    while trocas !=0:
        for i in range (len(lista)):
            for j in range (len(lista)):
                if lista[j] > lista[i]:
                    temp = lista[j]
                    lista[j] = lista[i]
                    lista[i]=temp
                    print(lista)
                    trocas=trocas+1
    
            trocas=0   
    return lista
           
lista=[1,7,8,4,5,2]

n= "1"
while n!="0":
    menu()

    n = input("Qual é a sua opção?:")
    if n == "1":
        lis = criarListaM()
        print(lis)
    elif n == "2":
        lis = criarListaA()
        print(lis)
    elif n == "3":
        lis = maiorElemento(lista)
        print(lis)
    elif n == "4":
        lis = bubbleSort(lista)
        print(lis)
    elif n == "0":
        print("Até à próxima")
    else:
        print("Opção inválida")
```
