O tpc4 consitiu na implementação de um progrma para trabalhar com frações, com o auxílio de tuplos e listas. 

Em termos práticos, o programa permite a criação e a simplificação de uma fração e fazer operaçóes envolvendo soma, subtração, divisão e multiplicação.

É também possível gerar uma lista de n frações, somar todos os valores da mesma e retornar a maior fração.

```python
from random import randint


def menu():
    print(""" Gestão para trabalhar com frações
          (1)-Criar fração
          (2)-Simplificar fração
          (3)-Somar
          (4)-Subtrair
          (5)-Multiplicar
          (6)-Dividir
          (7)-Gerar lista de n frações
          (8)-Somar Lista
          (9)-Maior fração da lista
          (0)-Sair
          """)


def criarFracao(numerador, denominador):
    return(numerador, denominador)


def verFracao(f):
    print(str(f[0])+ "/" + str(f[1]))


def mdc(a, b):
    if a < b:
        return mdc(b, a)
    elif a % b == 0:
        return b
    else:
        return mdc(a, a % b)


def simplificarFracao(f):
    numerador, denominador = f
    a = mdc(numerador, denominador)
    return (numerador/a, denominador/a)


def somarFrac(f1, f2):
    num1, denom1 = f1
    num2, denom2 = f2
    num = num1*denom2 + num2*denom1
    denom = denom1*denom2
    return simplificarFracao(criarFracao(num, denom))


def subtrairFrac(f1, f2):
    num1, denom1 = f1
    num2, denom2 = f2
    num = num1*denom2 - num2*denom1
    denom = denom1*denom2
    return simplificarFracao(criarFracao(num, denom))


def multiplicarFrac(f1, f2):
    num1, denom1 = f1
    num2, denom2 = f2
    num = num1*num2
    denom = denom1*denom2
    return simplificarFracao(criarFracao(num, denom))


def dividirFrac(f1, f2):
    num1, denom1 = f1
    num2, denom2 = f2
    num = num1*denom2
    denom = num2*denom1
    return simplificarFracao(criarFracao(num, denom))



def gerarLista():
    L = []
    l1 = int(input("Introduza o nª de elementos da lista:"))
    for i in range(l1):
        num = randint(0, 100)
        denom = randint(0, 100)
        L.append(criarFracao(num, denom))
    return L


def somarLista(L):
    res = (0, 1)
    for elem in L:
        res = somarFrac(res, elem)
    return (res)


def maiorFracaoLista(L):
    L1 = []
    for i in range(len(L)-1):
        num, denom = L[i]
        dec = int(num )/ int(denom)
        L1.append(dec)
    j = 0
    maxi=0
    while j <(len(L1)-1):
        if L1[j]<L1[j+1]:
             maxi=L1[j+1]   
        else:
            maxi=L1[j]           
        j=j+1
    ind= L1.index(maxi)
    return L[ind]

   


n = "1"
while n != 0:
    menu()

    n = int(input("Qual é a sua opção?:"))
    if n == 1:
        numerador= int(input("Introduza um numerador:"))
        denominador= int(input("Introduza um denominador:"))
        frac = criarFracao(numerador, denominador)
        print(frac)
    elif n == 2:
        simp = simplificarFracao(frac)
        print(simp)
    elif n == 3:
        num2= int(input("Introduza um numerador:"))
        denom2= int(input("Introduza um denominador:"))
        frac2=criarFracao(num2,denom2)
        somar = somarFrac(frac, frac2)
        print(somar)
    elif n == 4:
        sub = subtrairFrac(frac, frac2)
        print(sub)
    elif n == 5:
        mult = multiplicarFrac(frac, frac2)
        print(mult)
    elif n == 6:
        div = dividirFrac(frac, frac2)
        print(div)
    elif n == 7:
        ger = gerarLista()
        print(ger)
    elif n == 8:
        sl = somarLista(ger)
        print(sl)

    elif n == 9:
        maior = maiorFracaoLista(ger)
        print(maior)

    elif n == 0:
        print("Até à próxima")
    else:
        print("Opção inválida")
```
