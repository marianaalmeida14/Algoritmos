## Resumo:
Esta aplicação consiste num programa com funcões para criar polinómios a partir de uma lista e tuplos (termos), permitindo o cálculo do polinómio, a simplificação, a derivação e a visulaização numa tabela.

```python
def menu():
    print(""" App para polinómios
          (1) Criar Polinómio
          (2) Calcular Polinómio
          (3) Calcular Tabela
          (4) Simplificar Polinómio
          (5) Derivar Polinómio
          (0) Sair 
          """)


def definirTermo(coef, exp):
    return (coef, exp)


def criarPolinomio():
    return []


def inserirTermo(p, t):
    return p.append(t)


def verTermo(t):
    c, e = t
    if e == 0:
        return str(c)
    else:
        return (str(c) + "x^" + str(e))


def verPolinomio(p):
    res = ""
    for t in p:
        if res == "":
            res = verTermo(t)
        else:   
            res = res + " + " + verTermo(t)
    print(res)
        


def calcPol(p, x):
    soma = 0
    for t in p:
        c, e = t
        soma = soma + c * x**e
    return soma

def tabela(p, nlinhas):
    for i in range(nlinhas):
        print(str(i) + " :: " + str(calcPol(p, i)))

def chaveOrd(t):
    _, e = t
    return e


def ordenarPol(p):
    p.sort(reverse=True, key=chaveOrd)


def simplificarPol(p):
          ordenarPol(p)
          a,b=p[0]
          expmax=0
          L=[]
          polinomio=[]
          for i in range (expmax+1):
               for (c,e) in p:
                   if e==i:
                       L.append(c)
               soma=0
               for elem in L:
                    soma=soma+elem
               polinomio.append(definirTermo(soma, i))
               L=[]    
          ordenarPol(polinomio)
          print(polinomio)
          return polinomio


def derivarPol(p):
    res = []
    for (c,e) in p:
        if e > 0:
            res.append((c*e, e-1))
    return res



n = "1"
while n != 0:
    p=[]
    menu()
    n= int(input("Introduza a sua opção:"))

    if n == 1:
        n= int(input("Introduza o número de termos do polinomio:"))
        p= criarPolinomio()
        for i in range (n):
            coeficiente = input("Introduza o seu coeficiente :")
            expoente = int(input("Introduza o seu expoente :"))
            t =definirTermo(coeficiente, expoente)
            inserirTermo(p,t)
            verPolinomio(p)
        ordenarPol(p)
        verPolinomio(p)
            

    elif n == 2:
        if p==[]:
            n= int(input("Introduza o número de termos do polinomio:"))
            p= criarPolinomio()
            for i in range (n):
                coeficiente = input("Introduza o seu coeficiente :")
                expoente = int(input("Introduza o seu expoente :"))
                t =definirTermo(coeficiente, expoente)
                inserirTermo(p,t)
                verPolinomio(p)
            ordenarPol(p)
            verPolinomio(p)
            
            x = int(input("Introduza o valor de x:"))
            calc = calcPol(p, x)
            print(calc)

    elif n == 3:
        if p==[]:
            n= int(input("Introduza o número de termos do polinomio:"))
            p= criarPolinomio()
            for i in range (n):
                coeficiente = input("Introduza o seu coeficiente :")
                expoente = int(input("Introduza o seu expoente :"))
                t =definirTermo(coeficiente, expoente)
                inserirTermo(p,t)
                verPolinomio(p)
            ordenarPol(p)
            verPolinomio(p)
            nlinhas = int(input("Introduza o número de linhas da tabela:"))
            tab = tabela(p, nlinhas)
            print(tab)

    elif n == 4:
        if p==[]:
            n= int(input("Introduza o número de termos do polinomio:"))
            p= criarPolinomio()
            for i in range (n):
                coeficiente = input("Introduza o seu coeficiente :")
                expoente = int(input("Introduza o seu expoente :"))
                t =definirTermo(coeficiente, expoente)
                inserirTermo(p,t)
                verPolinomio(p)
            ordenarPol(p)
            verPolinomio(p)
            simp =simplificarPol(p)
            print(simp)

    elif n == 5:
        if p==[]:
            n= int(input("Introduza o número de termos do polinomio:"))
            p= criarPolinomio()
            for i in range (n):
                coeficiente = input("Introduza o seu coeficiente :")
                expoente = int(input("Introduza o seu expoente :"))
                t =definirTermo(coeficiente, expoente)
                inserirTermo(p,t)
                verPolinomio(p)
            ordenarPol(p)
            verPolinomio(p)
            
            der = simplificarPol(derivarPol(p))
            print(der)

    elif n == 0:
        print("Até à próxima")
    else:
        print("Opção inválida")
```
