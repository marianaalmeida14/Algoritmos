## Resumo: 
Esta aplicação permite a leitura da base de dados correspondente a informação de alunos, onde pode ser criada uma tabela que contêm o id, o nome, o curso e média de tpc de todos os alunos do ficheiro.
Com o auxílio do módulo matplotlib.pyplt é possível a visualização de um gráfico que indica quantos alunos há em cada curso e de outro que indica a média e número de alunos com essa média.
Foi desenvolvida a função top10 que dá como resultado uma lista com os alunos com as 10 médias mais altas: id, nome, curso, média e a função distribPorCurso e distribPorMedia.

```python
'''
BDAlunos=[aluno]
aluno=[id,nome,curso,[tpc1,tpc2,tpc3,tpc4]]
'''
import matplotlib.pyplot as plt


# exercicio 1
def getAluno(texto):
    textoLimpo=texto.replace("\"","")
    campos=textoLimpo.split(",")
    aluno=[]
    aluno.append(campos[0])
    aluno.append(campos[1])
    aluno.append(campos[2])
    aluno.append(campos[3:])
    lista=[]
    for s in aluno[3]:
        lista.append(int(s))
    aluno[3]=lista
    return aluno

# Leitura/carregamento da informação do ficheiro
def lerDataset(fnome):
    f=open(fnome, encoding="utf-8")
    bd=[]
    f.readline()
    for linha in f:
        bd.append(getAluno(linha))
    return bd

BD=[]
BD=lerDataset("alunos.csv")
print(BD)

# exercicio 2
def chaveOrd(a):
    return a[1]

# Listagem da informação
def listarDataset(bd):
    bd.sort(key=chaveOrd)
    print("id           nome            curso     média")
    print("--------------------------------------------")
    for a in bd:
        print(a[0] + " | " + a[1] + " | " + a[2] + " | " + str(sum(a[3])/4))

listarDataset(BD)

# exercico 3
def consultarDataset(bd,id):
    for a in bd:
        if a[0]== id:
            return print(a[0]+ "/" + a[1] + "/" + str(sum(a[3])/4))
        else:
            return print("Id não encontrado")



#Especifica um função que dá como resultado uma lista com os alunos com as 10 médias mais altas: id, nome, curso, média.
# exercício 4
def chaveOrdena(a):
    media =sum(a[3])/4
    return media

def top10(bd):
    bd.sort(key=chaveOrdena, reverse=True)
    BD2=[]
    for a in range(0,10):
        aluno2=[]
        for par in range (0,3) :
            aluno2.append(bd[a][par])
        aluno2.append((sum(bd[a][3])/4))
        BD2.append(aluno2)
    return BD2
    

# exercício 6
        
def distribPorMedia(bd):
    bd.sort(key=chaveOrdena)
    distribuicao2={}
    for a in bd:
            if round((sum (a[3])/4)) in distribuicao2.keys():
                distribuicao2[round((sum(a[3])/4))] += 1
            else:
                distribuicao2[round((sum(a[3])/4))] = 1
    return (distribuicao2)
                
    
 #exercício 5   
def distribPorCurso(bd):
    # Formato: ex.:[[LCC,23],[LEI,25],[MIEBIOM,22],[MIEFIS,20]]
    distribuicao={}
    for a in bd:
        if a[2] in distribuicao.keys():
            distribuicao[a[2]]+=1
        
        else:
            distribuicao[a[2]]=1
    return distribuicao

d=distribPorCurso(BD)
print (d)

#Especifica uma função que faz o plot dum gráfico com a distribuição de alunos por curso.
# EXERCICIO 7

def plotDistribPorCurso(bd):
    c=distribPorCurso(bd)
    x=c.keys()
    y=[]
    for i in x:
        y.append(c[i])
    plt.bar(x, y)
    plt.show()
    

# EXERCICIO 8

def plotDistribPorMedia(bd):
   cm=distribPorMedia(bd)
   x=cm.keys()
   y=[]
   for i in x:
        y.append(cm[i])
   plt.bar(x, y)
   plt.show()
    
    
BD = []
BD = lerDataset("alunos.csv")  
print(BD)
listarDataset(BD)
consultarDataset(BD, "a40")
#for linha in top10(BD):
    #print(linha)
print(top10(BD))
print(distribPorCurso(BD))
print(distribPorMedia(BD))
plotDistribPorCurso(BD)
plotDistribPorMedia(BD)      
```
